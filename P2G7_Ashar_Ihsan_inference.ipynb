```python
# import Libraries
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
```

```python
# load model
model = load_model('/content/drive/MyDrive/model_terbaiks.h5')

with open('classes.txt', 'r') as file:
    class_names = [line.strip() for line in file]
```

Dibawah kita akan melakukan testing model yang kita miliki menggunakan gambar yang bukan dari dataset kita. Tes prediksi ini dilakukan 3 kali dimana kita akan tes model apakah dapat memprediksi dengan baik 3 class berbeda.

## **Test Prediction 1**

```python
# Memuat gambar dan mengubah ukurannya sesuai dengan input model

img_path = '/content/sampah.jpeg'
img = image.load_img(img_path, target_size=(224, 224))  # Ukuran model kita 224 x 224
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)

plt.imshow(img)
plt.show()
```

```python
# Make prediction and get class name
predictions = model.predict(img_array)
predicted_class_index = np.argmax(predictions, axis=-1)[0]
predicted_class_name = class_names[predicted_class_index]

# Print the predicted class name
print("Prediksi kelas:", predicted_class_name)
```

Hasil tes pertama model kita berhasil memprediksi bahwa gambar merupakan garbage bag.
